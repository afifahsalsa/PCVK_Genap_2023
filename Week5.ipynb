# 1. Gamma Correction

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
print('Gamma Correction pada citra')
print('-------------------------------')
try:
  gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
  print('Error, not a number')
original = cv.imread('/content/drive/MyDrive/female.jpg')
gamma_img = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      gamma_img[y,x,c] = np.clip(255*pow(original[y,x,c]/255,(1/gamma)),0,255)

final_frame = cv.hconcat((original, gamma_img))
cv2_imshow(final_frame)
```

# 2. Image Depth

```python
from cv2.gapi import RGB2YUV
print('Image Depth pada citra')
print('-------------------------------')
try:
  bd = int(input('Masukkan nilai Bit Depth: '))
except ValueError:
  print('Error, not a number')

bit_depth=2*bd
level = 255/(pow(2,bit_depth)-1)
original = cv.imread('/content/drive/MyDrive/female.jpg', cv.IMREAD_GRAYSCALE)
depth_img = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
      depth_img[y,x] = np.clip(round(original[y, x]/level) * level,0,255);

final_frame = cv.hconcat((original, depth_img))
cv2_imshow(final_frame)
```

Jumlah citra di average 5

```python
import glob
from math import log10, sqrt
print('Average Denoising pada citra')
print('-------------------------------')
try:
  ad = int(input('Masukkan nilai : '))
except ValueError:
  print('Error, not a number')

def PSNR(ori, compressed):
  mse = np.mean((ori-compressed)**2)
  if(mse == 0):
    return 100
  max_pixel = 255.0
  psnr = 20*log10(max_pixel/sqrt(mse))
  return psnr

ori = cv.imread('/content/drive/MyDrive/galaxy.jpg')
denoising_avg = np.zeros(ori.shape, ori.dtype)

cv_img = []
for i in glob.glob('/content/100.jpg'):
  n = cv.imread(i)
  cv_img.append(n)
denoising_avg = np.mean(cv_img[:ad], axis=0).astype(np.uint8)

value = PSNR(ori, denoising_avg)
print(f"Nilai PSNR:{value} dB")

final_frame = cv.hconcat((ori, denoising_avg))
cv2_imshow(denoising_avg)
```

# 5. Image Masking menggunakan 'or', 'not', 'nand', 'xor'

1. OR

```python
img = cv.imread('/content/drive/MyDrive/couple.tiff')
mask = np.zeros(img.shape, dtype = np.uint8)
mask = cv.circle(mask, (70,50), 50, (255,255,255), -1)

mask2 = np.zeros(img.shape, dtype=np.uint8)
mask2 = cv.circle(mask, (190,50),50, (255,255,255),-1)

result = img | mask | mask2
#result[mask==0] = 255

final_frame = cv.hconcat((img, mask, result))
cv2_imshow(final_frame)
```

2. NOT

```python
img = cv.imread('/content/drive/MyDrive/couple.tiff')
mask = np.zeros(img.shape, dtype = np.uint8)
mask = cv.circle(mask, (70,50), 50, (255,255,255), -1)

mask2 = np.zeros(img.shape, dtype=np.uint8)
mask2 = cv.circle(mask, (190,50),50, (255,255,255),-1)

result = ~img

final_frame = cv.hconcat((img, mask, result))
cv2_imshow(final_frame)
```

3. NAND

```python
img = cv.imread('/content/drive/MyDrive/couple.tiff')
mask = np.zeros(img.shape, img.dtype)
cv.circle(mask, (70,50), 50, (255,255,255), -1)
cv.circle(mask, (190,50),50, (255,255,255),-1)

result = img & ~mask

final_frame = cv.hconcat((img, mask, result))
cv2_imshow(final_frame)
```

4. XOR

```python
img = cv.imread('/content/drive/MyDrive/couple.tiff')
mask = np.zeros(img.shape, img.dtype)
cv.circle(mask, (70,50), 50, (255,255,255), -1)
cv.circle(mask, (190,50),50, (255,255,255),-1)

result = img ^ mask

final_frame = cv.hconcat((img, mask, result))
cv2_imshow(final_frame)
```

Tuliskan hasil analisa anda:



*   Menggunakan OR akan menambahkan mask di citra
*   Menggunakan NOT akan menginverse warna gambar yang hasilnya akan menjadi negatif
*   Menggunakan NAND akan menambahkan mask dan membuat citra menjadi negative
*   Menggunakan XOR akan menutup gambar namun objek yang ditutup warnanya diinverse
